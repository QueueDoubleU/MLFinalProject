```python
def get_transcript(convo_id, remove_return=False):

  # pull conversation object
  convo = corpus.get_conversation(convo_id)

  # from conversation object, create list of utterance (i.e. the text bits) ids
  convo_utts = list(convo.iter_utterances())
  
  # combine text data from all utterances 
  if remove_return:
    convo_transcript = [utt.text.replace('\n', ' ') for utt in convo_utts]
  else:
    convo_transcript = [utt.text for utt in convo_utts]

  # join elements of list 
  convo_transcript = ''.join(convo_transcript)
  
  return convo_transcript

```

```python
def get_utt_transcript(utt_id, remove_return=False):

  # pull utterance object
  utt_obj = corpus.get_utterance(utt_id)

  # from utterance object, grab text
  utt_transcript = utt_obj.text
  
  return utt_transcript
```

```python
"""
Writes transcripts to folder path in the form {convo_id}_{case_id}
Resulting folder is roughly 1 GB. Tread lightly. 
"""

folder_path = '/Users/quinn/Documents/GradSchool/Spring2023/INFO251/Transcripts_uttid_caseid/'

for index, row in df_utt.iterrows():
    utt_id = row['utt_id']
    case_id = row['case_id']
    
    file_name = utt_id+'_'+case_id+'.txt'
    write_path = folder_path+file_name
    transcript = get_utt_transcript(utt_id)
    
    text_file = open(write_path, 'w')
    
    text_file.write(transcript)
    text_file.close()
```
